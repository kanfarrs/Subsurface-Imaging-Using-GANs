```python
from tensorflow.keras import backend
from tensorflow.compat.v1 import ConfigProto
from tensorflow.compat.v1 import InteractiveSession
from Code import utilities as ut
from Code import cycleGAN as model
import numpy as np
#config = ConfigProto()
#config.gpu_options.allow_growth = True
#session = InteractiveSession(config=config)
```

```python
#user input
batch_size = 5 #32
n_epoch = 200 #1
n_filters = 8 #8
comment = ''
saveModel = 'B' + str(batch_size) + '_FS' + str(n_filters) + '_E' + str(n_epoch) + comment #name of the saved model
```

```python
#data loading and processing
data = np.load('train_set_marm_100-30.npy')
focused, defocused = ut.dataProcess(data)
input_shape = (64, 64, 1)
```

```python
m = model.cycleGAN(input_shape, n_filters)
# generator, G: X -> Y
g_model_G = m.generator()
# generator, F: Y -> X
g_model_F = m.generator()
# discriminator: G -> [real/fake]
d_model_G = m.discriminator()
# discriminator: F -> [real/fake]
d_model_F= m.discriminator()
# composite: X -> Y -> [real/fake, X]
c_model_GtoF = m.composite_model(g_model_G, d_model_G, g_model_F, input_shape)
# composite: Y -> X -> [real/fake, Y]
c_model_FtoG = m.composite_model(g_model_F, d_model_F, g_model_G, input_shape)
```

```python
g_model_G.summary()
```

```python
m.train(d_model_G, d_model_F, g_model_G, g_model_F, c_model_GtoF, c_model_FtoG, defocused, focused, batch_size, n_epoch, saveModel)
```
